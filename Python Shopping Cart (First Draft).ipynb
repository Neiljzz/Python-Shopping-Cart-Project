# Python Shoppping Cart

Build a shopping cart program with the following capabilities:

1) Takes in an input
2) Stores user input into a dictionary or list
3) The User can add or delete items
4) The User can see current shopping list
5) The program Loops until user 'quits'
6) Upon quiting the program, prints out a receipt of the items with total and quantity.

```python
print("Welcome to the online Apple Store!")
shopping_cart = {}

def addproducts(product, num):
    if num.isdigit():
        num = int(num)
        if product in shopping_cart:
            shopping_cart[product] += num
        else:
            shopping_cart[product] = num


def delproducts(name, num):
    if name in shopping_cart:
        if num.isdigit():
            num = int(num)
            if num >= shopping_cart[name]:
                shopping_cart.pop(name)
            else:
                shopping_cart[name] -= num



while True:
    print("Please let us know what you want to do today: \n"
          "1. Add products 2. Delete products 3. Check out 4. End shopping")
    choice = input("Please enter the number: ")
    if choice in ["1", "2", "3", "4"]:
        if choice == "1":
            product_list = [
                ["IphoneX", "$799"],
                ["Macbook", "$1299"],
                ["IWatch", "$499"],
                ["MacMini", "$599"],
                ["Applepen", "$199"]
            ]
            print("Here are our products:")
            for product, price in enumerate(product_list, start=1):
                print(f"{product} {price}")

            index = input("Please enter the number of product you want to purchase (From 1-5):")
            if index.isdigit():
                index = int(index)
                if 0 <= index < len(product_list):
                    product = product_list[index-1][0]
                    num = input("How many you would like to buy?")
                    addproducts(product, num)
                    print(f"{num} {product} has been added to your cart!")
                else:
                    print("Invalid product number!")

        elif choice == "2":
            name = input("Please enter the product that you want to delete:")
            num = input("Please enter the number of product you want to delete:")
                delproducts(name, num)
            print("The product has been removed!")
            

        elif choice == "3":
            print("-----Check Out-----")
            print("Here are your products:")
            total = 0
            for product, num in shopping_cart.items():
                print(f"{product}: {num}")
                price = int(product_list[[p[0] for p in product_list].index(product)][1].strip("$"))
                total += num * price
            print(f"Your total is: ${total}")

        elif choice == "4":
            print("Thanks for shopping with us!")
            break
    else:
        print("Please try again")
```
